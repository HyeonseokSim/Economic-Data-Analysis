```python
import pandas as pd
import re

# 시간 매핑 딕셔너리 생성
def get_time_range(start_period, end_period):
    """주어진 교시 번호에 따라 시작 시간과 종료 시간을 매핑"""
    periods = {
        1: ('09:00', '09:30'), 2: ('09:30', '10:00'), 3: ('10:00', '10:30'), 4: ('10:30', '11:00'),
        5: ('11:00', '11:30'), 6: ('11:30', '12:00'), 7: ('12:00', '12:30'), 8: ('12:30', '13:00'),
        9: ('13:00', '13:30'), 10: ('13:30', '14:00'), 11: ('14:00', '14:30'), 12: ('14:30', '15:00'),
        13: ('15:00', '15:30'), 14: ('15:30', '16:00'), 15: ('16:00', '16:30'), 16: ('16:30', '17:00'),
        17: ('17:00', '17:30'), 18: ('17:30', '18:00'),
        19: ('18:15', '19:00'), 20: ('19:00', '19:45'), 21: ('19:45', '20:30'), 22: ('20:30', '21:15'),
        23: ('21:15', '22:00')
    }

    start_period = str(start_period).zfill(2)
    end_period = str(end_period).zfill(2)

    start_time, _ = periods[int(start_period)]
    _, end_time = periods[int(end_period)]

    return f"{start_time}~{end_time}"


# 건물명 매핑
building_keywords = [
    '경영', '상허관', '사', '예', '언어원', '법', '종강', '의', '생', '동', '산학', '수', '새', '건',
    '부', '문', '공A', '공B', '공C', '신공', '이', '창', '공D', '체육관', '연구동', '동L', '중장비', '공별', '농', '입', '교육'
]
mapped_names = [
    '경영관', '상허관', '교육과학관', '예술문화관', '언어교육관', '법학관', '법학관', '의생명과학연구관',
    '생명과학관', '동물생명과학관', '산학협동관', '수의학관', '새천년관', '건축관', '해봉부동산학관',
    '인문학관', 'A공학관', 'B공학관', 'C공학관', '신공학관', '과학관', '창의관', 'D공학관', '체육관',
    '연구동', '동물생명과학관', '중장비동', '공학관별관', '생명과학관부속동', '입학정보관', '교육과학관'
]
building_map = dict(zip(building_keywords, mapped_names))


def extract_info(row):
    # 입력값이 NaN일 경우 그대로 반환
    if pd.isna(row):
        return row, row, row, row, row

    # NaN이 아닌 경우 처리
    cleaned_row = re.sub(r'\(.*?\)', '', row)

    # 요일 추출
    days = re.findall(r'[월화수목금토일]', cleaned_row)

    # 일주일 중 수업 횟수 계산
    class_count = len(row.split(','))

    # 시간대
    times = re.findall(r'(\d{2})-(\d{2})', row)  # '04-06' 등 추출 (시작교시, 종료교시)
    mapped_times = [get_time_range(start, end) for start, end in times]

    # 소괄호 안의 건물명만 추출
    buildings = re.findall(r'\((.*?)\)', row)  # 소괄호 안의 문자열 추출
    mapped_buildings = []

    for building in buildings:
        # 건물명 매핑
        for keyword in building_keywords:
            if keyword in building:
                mapped_buildings.append(building_map[keyword])
                break

    # 층수 추출: 건물명 바로 뒤의 숫자 추출
    floor_numbers = []
    building_floor_pattern = r'(\D+)(\d)'  # 건물명 뒤에 있는 숫자 하나를 추출 (예: 공C487 -> 4)

    for building in buildings:
        match = re.search(building_floor_pattern, building)
        if match:
            floor_numbers.append(match.group(2))  # 두 번째 그룹: 숫자 (예: 4)

    return ', '.join(days), ', '.join(mapped_buildings), class_count, ', '.join(mapped_times), ', '.join(floor_numbers)

    
```

```python
# 새 컬럼 생성
df_22[['요일', '건물명', '일주일 중 수업횟수', '시간대', '층']] = df_22['강의요시'].apply(lambda x: pd.Series(extract_info(x)))
df_23[['요일', '건물명', '일주일 중 수업횟수', '시간대', '층']] = df_23['강의요시'].apply(lambda x: pd.Series(extract_info(x)))
df_24[['요일', '건물명', '일주일 중 수업횟수', '시간대', '층']] = df_24['강의요시'].apply(lambda x: pd.Series(extract_info(x)))
```

```python
def update_lecture_type(df):
    """
    '강의종류' 컬럼 값을 다음 조건에 따라 업데이트:
    1. '일반' -> '대면'
    2. 'e-러닝'이 포함된 값 -> 'e-러닝'
    3. 'B-러닝'이 포함된 값 -> 'B-러닝'
    """
    df['강의종류'] = df['강의종류'].replace(
        {
            '일반': '대면',
            **{v: 'e-러닝' for v in df['강의종류'].unique() if 'e-러닝' in str(v)},
            **{v: 'B-러닝' for v in df['강의종류'].unique() if 'B-러닝' in str(v)}
        }
    )
    return df

# df_22, df_23, df_24에 적용
df_22 = update_lecture_type(df_22)
df_23 = update_lecture_type(df_23)
df_24 = update_lecture_type(df_24)

# 결과 확인
print(df_22['강의종류'].unique())
print(df_23['강의종류'].unique())
print(df_24['강의종류'].unique())
```

# 점심시간대인지 아닌지 이진변수

```python
df_22_new['시간대'].head()
```

```python
def check_lunch_time_inclusion(time_ranges):
    # 점심시간 범위
    lunch_start = pd.to_datetime('12:00', format='%H:%M')
    lunch_end = pd.to_datetime('13:30', format='%H:%M')
    
    # 시간대를 ','로 분리
    for time_range in str(time_ranges).split(','):
        try:
            # '~'로 분리하여 시작 시간과 종료 시간 추출
            start_time, end_time = time_range.strip().split('~')
            start_time = pd.to_datetime(start_time, format='%H:%M')
            end_time = pd.to_datetime(end_time, format='%H:%M')
            
            # 점심시간 범위와 교차하는지 확인
            if (start_time < lunch_end) and (end_time > lunch_start):
                return 1  # 점심시간을 포함하면 1 반환
        except ValueError:
            # 시간대 형식이 올바르지 않은 경우 예외 처리
            continue
    
    # 점심시간 포함되지 않으면 0 반환
    return 0

# '점심시간 포함 개수' 컬럼 생성
df_22_new['점심시간 포함 개수'] = df_22_new['시간대'].apply(check_lunch_time_inclusion)
df_23_new['점심시간 포함 개수'] = df_23_new['시간대'].apply(check_lunch_time_inclusion)
df_24_new['점심시간 포함 개수'] = df_24_new['시간대'].apply(check_lunch_time_inclusion)
```
